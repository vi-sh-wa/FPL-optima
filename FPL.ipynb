```python
import requests
import pandas as pd
```

```python
url = "https://fantasy.premierleague.com/api/bootstrap-static/"
```

```python
response = requests.get(url)
data = response.json()
```

```python
team_data = data['teams']
player_data = data['elements']
player_stats = data['element_stats']
player_types = data['element_types']
```
